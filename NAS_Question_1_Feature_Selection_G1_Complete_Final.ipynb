# Missing Values

```python
subject['Maths %'].fillna(subject['Maths %'].mean(), inplace=True)
subject['Reading %'].fillna(subject['Reading %'].mean(), inplace=True)
subject['Science %'].fillna(subject['Science %'].mean(), inplace=True)
subject['Social %'].fillna(subject['Social %'].mean(), inplace=True)

subject.head()
```

```python
subject.max()
```

```python
subject['Percent'] =((subject['Maths %'] + subject['Reading %'] + subject['Science %'] + subject['Social %'])/400)*100
```

```python
subject.head()
```

```python
math    = subject["Maths %"]
reading = subject["Reading %"]
science = subject["Science %"]
social  = subject["Social %"]
percent = ((math+reading+science+social)/400) * 100

max_mark = np.max(subject['Percent'])
poor_mark = np.min(subject['Percent'])
avg_mark  = np.average(subject['Percent'])

Threshold = max_mark-avg_mark
super_threshold_indices = percent > Threshold
a = np.copy(percent)
a[super_threshold_indices] = 1
a[~super_threshold_indices]= 0
```

```python
Class     = pd.DataFrame(a,columns=["Class"])
```

```python
# define X and y and change it to 'float'
```

```python
X = data
y = Class
```

# 1. Find Important Features using Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
y = np.ravel(y)
model = RandomForestRegressor()
model.fit(X,y)
```

```python
model.feature_importances_
srt =pd.Series(model.feature_importances_,index=X.columns)
```

```python
srt = srt.sort_values(axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last')
```

```python
srt.plot(kind='barh', figsize=(15,20))
plt.title('Importances Features')
```

# 2. Find Important Feature using ExtraTreesClassifier

```python
y = np.ravel(y)
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X, y)
feature_importance=model.feature_importances_
```

```python
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0])
plt.rcParams['figure.figsize'] = [15,15]
plt.barh(pos, feature_importance[sorted_idx])
plt.yticks(pos, X.columns[sorted_idx])
plt.title('Importances Features')
plt.show()
```
